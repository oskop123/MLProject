```python
import pandas as pd
import numpy as np
import random
```

```python
N = 5

data = np.empty((10**5, N + 1))

for x in range(data.shape[0]):
  for y in range(data.shape[1]):
      data[x,y] = random.random()

data_frame = pd.DataFrame(data, columns=('a%d' % (i) for i in range(N+1)))
```
