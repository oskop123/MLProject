```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow import keras
```

```python
X = pd.read_pickle('/content/drive/MyDrive/Colab Notebooks/ML/Projekt/features.pkl')
Y = pd.read_pickle('/content/drive/MyDrive/Colab Notebooks/ML/Projekt/labels.pkl')
```

```python
X_np = X.to_numpy()
Y_np = Y.to_numpy()

DEGREE = X_np.shape[1] - 1
```

```python
x_train = X_np[:80000]
y_train = Y_np[:80000]

x_val = X_np[80000:90000]
y_val = Y_np[80000:90000]

x_test = X_np[90000:]
y_test = Y_np[90000:]

print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
print(x_test.shape, y_test.shape)
```

```python
model = keras.Sequential()

x = 250

model.add(layers.Dense(x, activation="relu", input_shape=(DEGREE+1,)))
model.add(layers.Dropout(.2))
model.add(layers.Dense(x, activation="tanh"))
model.add(layers.Dropout(.2))
model.add(layers.Dense(x, activation="relu"))
model.add(layers.Dropout(.2))
model.add(layers.Dense(x, activation="tanh"))
model.add(layers.Dense(DEGREE*2))

model.summary()
```

```python
BATCH_SIZE = 256

model.compile(
    optimizer=keras.optimizers.Adam(),
    loss=keras.losses.MeanAbsoluteError(),
    metrics=[keras.metrics.MeanAbsoluteError()],
)
```

```python
model.fit(x_train, y_train, batch_size=BATCH_SIZE, epochs=20, verbose=1, validation_data=(x_val, y_val))
```

```python
model.fit(x_train, y_train, batch_size=BATCH_SIZE, epochs=20, verbose=1, validation_data=(x_val, y_val))
```

```python
# Evaluate the model on the test data using `evaluate`
print("Evaluate on test data")
results = model.evaluate(x_test, y_test, batch_size=256, verbose=2)
print("test loss, test acc:", results)
```
